```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Book_multisources_clean.csv")
```

Histogram of book prices (the distribution of book prices (where most books fall in terms of price))

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=30, kde=True, color="#4575b4")  
plt.title("Distribution of Book Prices")
plt.xlabel("Price ($)")
plt.ylabel("Number of Books")
plt.show()
```

Boxplot of Prices by Category (Top 10 Categories)
"which categories have higher/lower prices and highlights outliers"

```python
top10 = df["Category"].value_counts().nlargest(10).index
plt.figure(figsize=(10,6))
sns.boxplot(
    x="Category",
    y="Price",
    data=df[df["Category"].isin(top10)],
    palette=sns.color_palette("RdBu", 10)
)
plt.title("Book Prices by Category (Top 10)")
plt.xticks(rotation=45)
plt.xlabel("Category")
plt.ylabel("Price ($)")
plt.show()
```

Scatter Plot (Price vs Rating)

```python
plt.figure(figsize=(10,7))
sns.scatterplot(
    x="Rating",
    y="Price",
    data=df,
    hue="Availability",
    palette=["#d73027", "#4575b4"],  
    alpha=0.7
)
plt.title("Price vs Rating by Availability")
plt.xlabel("Rating")
plt.ylabel("Price ($)")
plt.show()
```

Countplot of Ratings (the number of books for each rating (1–5).)

```python
plt.figure(figsize=(7,5))
sns.countplot(
    x="Rating",
    data=df,
    palette=sns.color_palette("RdBu", df["Rating"].nunique())
)
plt.title("Count of Books per Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Books")
plt.show()
```

Pie Chart of Availability (proportion of available vs unavailable books)

```python
availability_counts = df["Availability"].value_counts()
plt.figure(figsize=(6,6))
colors = ["#d73027", "#4575b4"]  
plt.pie(
    availability_counts,
    labels=["Available","Not Available"],
    autopct="%1.1f%%",
    colors=colors
)
plt.title("Book Availability")
plt.show()
```

Heatmap of Correlation (relationships between numeric variables (Price, Rating, Availability, etc.))

```python
plt.figure(figsize=(7,5))
corr = df[["Price","Availability","Rating","Price_Normalized"]].corr()
sns.heatmap(corr, annot=True, cmap="RdBu", fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```
